1.สร้าง console application project

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
BaseClass BC = new DerivedClass();
BC.A(); 

abstract class BaseClass
{
    abstract public void A();
}
class DerivedClass : BaseClass
{
    public override void A()
    {
        System.Console.WriteLine("Implementation of inheritance classes");
    }
}
```

```cmd
dotnet build  Lab12_Ex01
```

การใช้ abstract class ในกรณีนี้ช่วยให้คลาสที่สืบทอด (เช่น DerivedClass) ต้อง implement เมธอด A() ผลลัพธ์ที่ได้จากการเรียกเมธอด A() จะเป็นของคลาสที่มีการ implement หรือ override เมธอดนั้น (ในที่นี้คือ DerivedClass)
